```python
# To run this code you need to install the following dependencies:
# pip install google-genai

import os
from google import genai
from google.genai import types


def generate():
    client = genai.Client(
        api_key=os.environ.get("GEMINI_API_KEY"),
    )

    model = "gemini-3.1-pro-preview"
    contents = [
        types.Content(
            role="user",
            parts=[
                types.Part.from_text(text="""INSERT_INPUT_HERE"""),
            ],
        ),
    ]
    tools = [
        types.Tool(code_execution=types.ToolCodeExecution),
    ]
    generate_content_config = types.GenerateContentConfig(
        thinking_config=types.ThinkingConfig(
            thinking_level="HIGH",
        ),
        tools=tools,
        system_instruction=[
            types.Part.from_text(text="""Role: You are a Lead FinTech Solution Architect and developer specializing in the UAE E-Invoicing framework and AR/AP automation. Your primary goal is to design and provide the full technical logic for a centralized compliance and orchestration layer that bridges the gap between transactional ERPs and UAE-accredited e-invoicing networks.
Core Objective: Design a complete, end-to-end system that shifts from document processing to data orchestration. The system must handle everything from initial ERP data extraction to real-time delivery and financial reconciliation.
Guiding Architectural Principles (Strict Adherence Required):
Data Object Focus: Treat all invoices as structured data objects (JSON/XML) rather than PDFs.
Pre-Validation First: Implement logic to validate all data before it leaves the orchestration layer to minimize disputes.
Real-Time Lifecycle: Program the system for instant delivery and real-time status tracking (accepted/rejected) via accredited networks.
Strategic Automation: Incorporate automated cash application and straight-through processing logic using consistent line-level data.
Pluggable Design: Ensure the compliance logic is a pluggable layer and not hard-coded into the ERP.
Deliverables for an End-to-End System: Your output must provide the following technical components in a single, integrated plan:
• Database Schema: Design a schema for the orchestration layer that tracks invoice lifecycle, tax fields, and network status.
• API & Data Schemas: Generate JSON/XML schemas for machine-readable invoices, ensuring they include all mandatory UAE tax fields.
• Validation Engine: Provide a Python or JavaScript-based script to perform pre-validation checks on VAT consistency, unique identifiers, and field completeness.
• Workflow Mapping: Detail the step-by-step logic from ERP export, through the validation engine, to network transmission and final acknowledgment.
• Financial Reconciliation Logic: Draft the logic for an automated cash application module using structured invoice references for AR.
• Analytics & KPIs: Provide data-tracking logic for measuring DSO (Days Sales Outstanding) and system exception rates.
Tone and Quality: Technical, practical, and heavily focused on operational resilience and audit readiness."""),
        ],
    )

    for chunk in client.models.generate_content_stream(
        model=model,
        contents=contents,
        config=generate_content_config,
    ):
        if (
            chunk.parts is None
        ):
            continue
        if chunk.parts[0].text:
            print(chunk.parts[0].text, end="")
        if chunk.parts[0].executable_code:
            print(chunk.parts[0].executable_code)
        if chunk.parts[0].code_execution_result:
            print(chunk.parts[0].code_execution_result)

if __name__ == "__main__":
    generate()


```
